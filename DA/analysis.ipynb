```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
amd_laptop_1024 = pd.read_csv('AMD_laptop_1024.csv')
amd_laptop_1024['expInfo'] = '1024_AMD_laptop'
amd_laptop_1024
```

```python

%matplotlib inline
# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot branch instructions
axs[0].bar(amd_laptop_1024['Method'], amd_laptop_1024['Branches'])
axs[0].set_xlabel('Method')
axs[0].set_ylabel('Branch Instructions')
axs[0].set_title('Branch Instructions for Different Methods')
axs[0].tick_params(axis='x', rotation=45)

# Plot branch misses
axs[1].bar(amd_laptop_1024['Method'], amd_laptop_1024['Branch Misses'])
axs[1].set_xlabel('Method')
axs[1].set_ylabel('Branch Misses')
axs[1].set_title('Branch Misses for Different Methods')
axs[1].tick_params(axis='x', rotation=45)

# Plot cache misses
axs[2].bar(amd_laptop_1024['Method'], amd_laptop_1024['L1-dcache Load Misses'])
axs[2].set_xlabel('Method')
axs[2].set_ylabel('Cache Misses')
axs[2].set_title('Cache Misses for Different Methods')
axs[2].tick_params(axis='x', rotation=45)

# Add a title to the entire figure
fig.suptitle('AMD laptop 1024')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Define bar width
bar_width = 0.2

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Create x-axis positions
methods = amd_laptop_1024['Method']
x = np.arange(len(methods))  # Base x locations for bars

# Compute branch miss ratio
amd_laptop_1024_ratio = amd_laptop_1024['Branch Misses'] / amd_laptop_1024[
    'Branches']
amd_server_1024_ratio = amd_server_1024['Branch Misses'] / amd_server_1024[
    'Branches']
intel_1024_ratio = intel_1024['Branch Misses'] / intel_1024['Branches']

# Plot Branch Miss Ratio for 1024 matrix
bars1 = axs[0].bar(x - bar_width,
                   amd_laptop_1024_ratio,
                   width=bar_width,
                   label='AMD Laptop 1024')
bars2 = axs[0].bar(x,
                   amd_server_1024_ratio,
                   width=bar_width,
                   label='AMD Server 1024',
                   alpha=0.7)
bars3 = axs[0].bar(x + bar_width,
                   intel_1024_ratio,
                   width=bar_width,
                   label='Intel 1024',
                   alpha=0.5)

# Add labels to bars
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        axs[0].text(bar.get_x() + bar.get_width() / 2,
                    height,
                    f'{height:.2%}',
                    ha='center',
                    va='bottom',
                    fontsize=10)

axs[0].set_xticks(x)
axs[0].set_xticklabels(methods, rotation=45)
axs[0].set_xlabel('Method')
axs[0].set_ylabel('Branch Miss Ratio')
axs[0].set_title('Branch Miss Ratio for Different Methods (1024 Matrix)')
axs[0].legend()

# Create x-axis positions for 4096 matrix
methods_4096 = amd_laptop_4096['Method']
x_4096 = np.arange(len(methods_4096))

# Compute branch miss ratio for 4096 matrix
amd_laptop_4096_ratio = amd_laptop_4096['Branch Misses'] / amd_laptop_4096[
    'Branches']
amd_server_4096_ratio = amd_server_4096['Branch Misses'] / amd_server_4096[
    'Branches']
intel_4096_ratio = intel_4096['Branch Misses'] / intel_4096['Branches']

# Plot Branch Miss Ratio for 4096 matrix
bars4 = axs[1].bar(x_4096 - bar_width,
                   amd_laptop_4096_ratio,
                   width=bar_width,
                   label='AMD Laptop 4096')
bars5 = axs[1].bar(x_4096,
                   amd_server_4096_ratio,
                   width=bar_width,
                   label='AMD Server 4096',
                   alpha=0.7)
bars6 = axs[1].bar(x_4096 + bar_width,
                   intel_4096_ratio,
                   width=bar_width,
                   label='Intel 4096',
                   alpha=0.5)

# Add labels to bars
for bars in [bars4, bars5, bars6]:
    for bar in bars:
        height = bar.get_height()
        axs[1].text(bar.get_x() + bar.get_width() / 2,
                    height,
                    f'{height:.2%}',
                    ha='center',
                    va='bottom',
                    fontsize=10)

axs[1].set_xticks(x_4096)
axs[1].set_xticklabels(methods_4096, rotation=45)
axs[1].set_xlabel('Method')
axs[1].set_ylabel('Branch Miss Ratio')
axs[1].set_title('Branch Miss Ratio for Different Methods (4096 Matrix)')
axs[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
bar_width = 0.2
methods = amd_laptop_1024['Method']

# Get X axis index
x = np.arange(len(methods))
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# 1024 Matrix
axs[0].bar(x - bar_width, (amd_laptop_1024['Task Clock (msec)'] / 1000) /
           amd_laptop_1024['Elapsed Time (s)'],
           width=bar_width,
           label='AMD Laptop 1024')
axs[0].bar(x, (amd_server_1024['Task Clock (msec)'] / 1000) /
           amd_server_1024['Elapsed Time (s)'],
           width=bar_width,
           label='AMD Server 1024',
           alpha=0.7)
axs[0].bar(x + bar_width, (intel_1024['Task Clock (msec)'] / 1000) /
           intel_1024['Elapsed Time (s)'],
           width=bar_width,
           label='Intel 1024',
           alpha=0.5)

axs[0].set_xticks(x)
axs[0].set_xticklabels(methods, rotation=45)
axs[0].set_xlabel('Method')
axs[0].set_ylabel('CPU core utilization')
axs[0].set_title('CPU core utilization (1024 Matrix)')
axs[0].legend()
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Add labels
for i, v in enumerate((amd_laptop_1024['Task Clock (msec)'] / 1000) /
                      amd_laptop_1024['Elapsed Time (s)']):
    axs[0].text(i - bar_width, v + 0.05, f'{v:.2f}', ha='center')
for i, v in enumerate((amd_server_1024['Task Clock (msec)'] / 1000) /
                      amd_server_1024['Elapsed Time (s)']):
    axs[0].text(i, v + 0.05, f'{v:.2f}', ha='center')
for i, v in enumerate(
    (intel_1024['Task Clock (msec)'] / 1000) / intel_1024['Elapsed Time (s)']):
    axs[0].text(i + bar_width, v + 0.05, f'{v:.2f}', ha='center')

# 4096 Matrix
axs[1].bar(x - bar_width, (amd_laptop_4096['Task Clock (msec)'] / 1000) /
           amd_laptop_4096['Elapsed Time (s)'],
           width=bar_width,
           label='AMD Laptop 4096')
axs[1].bar(x, (amd_server_4096['Task Clock (msec)'] / 1000) /
           amd_server_4096['Elapsed Time (s)'],
           width=bar_width,
           label='AMD Server 4096',
           alpha=0.7)
axs[1].bar(x + bar_width, (intel_4096['Task Clock (msec)'] / 1000) /
           intel_4096['Elapsed Time (s)'],
           width=bar_width,
           label='Intel 4096',
           alpha=0.5)

axs[1].set_xticks(x)
axs[1].set_xticklabels(methods, rotation=45)
axs[1].set_xlabel('Method')
axs[1].set_ylabel('CPU core utilization')
axs[1].set_title('CPU core utilization (4096 Matrix)')
axs[1].legend()
axs[1].grid(axis='y', linestyle='--', alpha=0.6)

# Add labels
for i, v in enumerate((amd_laptop_4096['Task Clock (msec)'] / 1000) /
                      amd_laptop_4096['Elapsed Time (s)']):
    axs[1].text(i - bar_width, v + 0.05, f'{v:.2f}', ha='center')
for i, v in enumerate((amd_server_4096['Task Clock (msec)'] / 1000) /
                      amd_server_4096['Elapsed Time (s)']):
    axs[1].text(i, v + 0.05, f'{v:.2f}', ha='center')
for i, v in enumerate(
    (intel_4096['Task Clock (msec)'] / 1000) / intel_4096['Elapsed Time (s)']):
    axs[1].text(i + bar_width, v + 0.05, f'{v:.2f}', ha='center')

# 调整布局
plt.tight_layout()
plt.show()
```
