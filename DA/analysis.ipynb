```python
import pandas as pd
```

```python
amd_laptop_1024 = pd.read_csv('AMD_laptop_1024.csv')
amd_laptop_1024['expInfo'] = '1024_AMD_laptop'
amd_laptop_1024
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot branch instructions
axs[0].bar(amd_laptop_1024['Method'], amd_laptop_1024['Branches'])
axs[0].set_xlabel('Method')
axs[0].set_ylabel('Branch Instructions')
axs[0].set_title('Branch Instructions for Different Methods')
axs[0].tick_params(axis='x', rotation=45)

# Plot branch misses
axs[1].bar(amd_laptop_1024['Method'], amd_laptop_1024['Branch Misses'])
axs[1].set_xlabel('Method')
axs[1].set_ylabel('Branch Misses')
axs[1].set_title('Branch Misses for Different Methods')
axs[1].tick_params(axis='x', rotation=45)

# Plot cache misses
axs[2].bar(amd_laptop_1024['Method'], amd_laptop_1024['L1-dcache Load Misses'])
axs[2].set_xlabel('Method')
axs[2].set_ylabel('Cache Misses')
axs[2].set_title('Cache Misses for Different Methods')
axs[2].tick_params(axis='x', rotation=45)

# Add a title to the entire figure
fig.suptitle('AMD laptop 1024')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Define bar width
bar_width = 0.2

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Plot L1-dcache Load Misses for AMD 1024 and Cache Misses for Intel 1024
axs[0].bar(amd_laptop_1024['Method'], amd_laptop_1024['L1-dcache Load Misses'], width=bar_width, label='AMD Laptop 1024')
axs[0].bar(amd_server_1024['Method'], amd_server_1024['L1-dcache Load Misses'], width=bar_width, label='AMD Server 1024', alpha=0.7)
axs[0].bar(intel_1024['Method'], intel_1024['Cache Misses'], width=bar_width, label='Intel 1024', alpha=0.5)
axs[0].set_xlabel('Method')
axs[0].set_ylabel('L1-dcache Load Misses / Cache Misses')
axs[0].set_title('L1-dcache Load Misses / Cache Misses for Different Methods (1024 Matrix)')
axs[0].tick_params(axis='x', rotation=45)
axs[0].legend()

# Plot L1-dcache Load Misses for AMD 4096 and Cache Misses for Intel 4096
axs[1].bar(amd_laptop_4096['Method'], amd_laptop_4096['L1-dcache Load Misses'], width=bar_width, label='AMD Laptop 4096')
axs[1].bar(amd_server_4096['Method'], amd_server_4096['L1-dcache Load Misses'], width=bar_width, label='AMD Server 4096', alpha=0.7)
axs[1].bar(intel_4096['Method'], intel_4096['Cache Misses'], width=bar_width, label='Intel 4096', alpha=0.5)
axs[1].set_xlabel('Method')
axs[1].set_ylabel('L1-dcache Load Misses / Cache Misses')
axs[1].set_title('L1-dcache Load Misses / Cache Misses for Different Methods (4096 Matrix)')
axs[1].tick_params(axis='x', rotation=45)
axs[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Plot Task Clock per Elapsed Time for AMD 1024 and Intel 1024
axs[0].bar(amd_laptop_1024['Method'], (amd_laptop_1024['Task Clock (msec)'] / amd_laptop_1024['Elapsed Time (s)']) * 100, width=bar_width, label='AMD Laptop 1024')
axs[0].bar(amd_server_1024['Method'], (amd_server_1024['Task Clock (msec)'] / amd_server_1024['Elapsed Time (s)']) * 100, width=bar_width, label='AMD Server 1024', alpha=0.7)
axs[0].bar(intel_1024['Method'], (intel_1024['Task Clock (msec)'] / intel_1024['Elapsed Time (s)']) * 100, width=bar_width, label='Intel 1024', alpha=0.5)
axs[0].set_xlabel('Method')
axs[0].set_ylabel('# of utilized CPU')
axs[0].set_title('CPU core utilization (1024 Matrix)')
axs[0].tick_params(axis='x', rotation=45)
axs[0].legend()

# Plot Task Clock per Elapsed Time for AMD 4096 and Intel 4096
axs[1].bar(amd_laptop_4096['Method'], (amd_laptop_4096['Task Clock (msec)'] / amd_laptop_4096['Elapsed Time (s)']) * 100, width=bar_width, label='AMD Laptop 4096')
axs[1].bar(amd_server_4096['Method'], (amd_server_4096['Task Clock (msec)'] / amd_server_4096['Elapsed Time (s)']) * 100, width=bar_width, label='AMD Server 4096', alpha=0.7)
axs[1].bar(intel_4096['Method'], (intel_4096['Task Clock (msec)'] / intel_4096['Elapsed Time (s)']) * 100, width=bar_width, label='Intel 4096', alpha=0.5)
axs[1].set_xlabel('Method')
axs[1].set_ylabel('Task Clock per Elapsed Time (%)')
axs[1].set_title('CPU core utilization (4096 Matrix)')
axs[1].tick_params(axis='x', rotation=45)
axs[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
